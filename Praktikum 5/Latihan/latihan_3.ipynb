```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
y3 = np.array([4, 2, 4, 12])
# Membuat plot untuk garis y1 dan y2
# plt.plot(y1, label='Garis 1')
# plt.plot(y2, label='Garis 2')
#modifikasi
plt.plot(y1, label="Grafik 1", marker="H", color="green", ls="dashed")
plt.plot(y2, label="Grafik 2", color="yellow")
plt.plot(y3, label="Grafik 3", marker="*", ls="dotted", color="purple")

# Menambahkan judul dan label sumbu
plt.title('Plot Dua Garis')
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')

# Menambahkan keterangan (legend)
plt.legend()
plt.show()
```
